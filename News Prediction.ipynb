```python
#In this project I built three different models to predict if a given news is fake or real. 
#For each model I used different strategies and then I compared their accuracy.

#importing libraries
import pandas as pd
import numpy as np
import itertools
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#getting the data
news = pd.read_csv(r"C:\Users\ponce\Downloads\Data Sc\Python\news.csv")
news
```

```python
#Formating the data 
del news["Unnamed: 0"]
news.dropna()
news['label'] = news['label'].map({'REAL': 1, 'FAKE': 0}) # Real =1, FAKE = 0
news
```

```python
#Creating the models

model =PassiveAggressiveClassifier(max_iter=50)
model_2 = DecisionTreeClassifier()
model_3 = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)

labels = news["label"]

#splitting the data for training and test
x_train, x_test, y_train, y_test = train_test_split(news['text'], 
                                                    labels,test_size=0.2, 
                                                    random_state=9)

#Vectorizing the data
tfidf_vectorizer=TfidfVectorizer(stop_words='english', max_df=0.7)
tfidf_train=tfidf_vectorizer.fit_transform(x_train)
tfidf_test=tfidf_vectorizer.transform(x_test)

#fitting the model
model.fit(tfidf_train,y_train)
model_2.fit(tfidf_train,y_train)
model_3.fit(tfidf_train,y_train)

#Using the test data to make predictions with each model
prediction = model.predict(tfidf_test)
prediction_2 = model_2.predict(tfidf_test)
prediction_3 = model_3.predict(tfidf_test)

#Printing the accuracy of each model
print("The Passive Aggressive Classifier Accuracy: "+str(accuracy_score(y_test,prediction)))
print("The Decision Tree Classifier Accuracy: "+str(accuracy_score(y_test,prediction_2)))
print("The Random Forest Classifier Accuracy: "+str(accuracy_score(y_test,prediction_3)))
```
